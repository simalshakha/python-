```python


import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
n_samples = 1000

seed = 30
```

```python
noisy_circles = datasets.make_moons(n_samples=n_samples, noise=0.05, random_state=seed)
```

```python
# noisy_circles
```

```python
X, y = noisy_circles
X.shape
```

```python
y.shape
```

```python
plt.scatter(
    X[:, 0], 
    X[:, 1], 
    s=5,
    c=list(
        map(
            lambda x: "black",
            y
        )
    )
)
```

```python
plt.scatter(
    X[:, 0], 
    X[:, 1], 
    s=5,
    c=list(
        map(
            lambda x: "red" if x == 1 else "blue",
            y
        )
    )
)
```

### Scaling the X varaible

```python
X = StandardScaler().fit_transform(X)
```

```python
X.shape
```

### Clustering using KMeans Algorithm

```python
kmeans_cluster_predict = KMeans(n_clusters=2).fit_predict(X)
kmeans_cluster_predict[:100]
```

```python
plt.scatter(
    X[:, 0], 
    X[:, 1], 
    s=5,
    c=list(
        map(
            lambda x: "red" if x == 0  else "blue" , 
            kmeans_cluster_predict
        )
    )
)
```

```python
dbscan_cluster = DBSCAN(min_samples=6, eps=0.25).fit_predict(X)
dbscan_cluster[:100]
```

```python
plt.scatter(
    X[:, 0], 
    X[:, 1],
    s=5,
    c=list(
        map(
            lambda x: "red" if x == 0 else "blue" if x == 1 else "black",
            dbscan_cluster
        )
    )
)
```
