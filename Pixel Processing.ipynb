```python
import os
import numpy as np
import cv2
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
rabbit_img = cv2.imread('rabbit.jpeg')
```

```python
rabbit_gray = cv2.cvtColor(rabbit_img, cv2.COLOR_BGR2GRAY)
```

```python
rabbit_contrast_img = cv2.convertScaleAbs(rabbit_gray, alpha=1, beta=-100)
```

```python
hist = cv2.calcHist([rabbit_contrast_img], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
cv2.imshow('Rabbit', rabbit_gray)
cv2.imshow('Rabbit Contast', rabbit_contrast_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Histogram Equlization

```python
lenna_img = cv2.imread('Lenna.png')
```

```python
lenna_gray = cv2.cvtColor(lenna_img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(lenna_gray.ravel())
```

```python
hist = cv2.calcHist([lenna_gray], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
plt.plot(hist.cumsum())
```

```python
lenna_gray_eq = cv2.equalizeHist(lenna_gray)
```

```python
hist = cv2.calcHist([lenna_gray_eq], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
plt.plot(hist.cumsum())
```

```python
cv2.imshow('Lenna Original', lenna_img)
cv2.imshow('Lenna Gray', lenna_gray)
cv2.imshow('Lenna Gray EQ', lenna_gray_eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
