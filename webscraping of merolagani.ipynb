```python
url = 'https://merolagani.com/LatestMarket.aspx'  # Example URL, replace with your target
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
s= soup.select('div#ctl00_ContentPlaceHolder1_LiveTrading.table-responsive')
table = s[0].find('table')
table
```

```python
header = table.find_all('th')
header = [i.text.strip() for i in header]

header=header[:9]
header[3],header[4]=header[4],header[3]
header[4],header[5]=header[5],header[4]
header
```

```python
body= table.find_all('td')
u = 'https://merolagani.com'
a=[]
for td in body:
    if td.find('a'):
        a.append(td.find('a'))
s = [tag for i, tag in enumerate(a) if i % 3 == 0]
href_values = [tag.get('href') for tag in s]
name= [i.text.strip() for i in s]
href_values=[u+values for  values in href_values]
result_dict = {k: v for k, v in zip(name, href_values)}
result_dict
href_values
```

```python
body= table.find_all('td')
body= [i.text.strip() for i in body]
body=header+body
body
```

```python
cleaned_data = [value.strip().replace('"', '').replace(',', '') for value in body]
rows=[]

for i in range(0, len(cleaned_data), 9):
    sublist = cleaned_data[i:i + 9]
    rows.append(sublist)
newrow=[]
for row in rows:
    row=row[:7]
    newrow.append(row)

newrow
```

```python
filename = "output.csv"

with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(newrow)
```
